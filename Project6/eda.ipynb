```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# loading the file
df = pd.read_csv("superconductivty+data/train.csv")

print("Shape of dataset:", df.shape)
df.head()
```

```python
df.info()

# checking missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['critical_temp'], bins=50, kde=True)
plt.title("Distribution of Critical Temperature")
plt.xlabel("Critical Temperature (K)")
plt.ylabel("Count")
plt.show()
```

```python
corr_matrix = df.corr()

top_corr_features = corr_matrix['critical_temp'].abs().sort_values(ascending=False)[1:16]

plt.figure(figsize=(12, 8))
sns.heatmap(df[top_corr_features.index.tolist() + ['critical_temp']].corr(), annot=True, cmap='coolwarm')
plt.title("Top 15 Correlated Features with Critical Temperature")
plt.show()

# top 15 features most correlated with critical_temp
```

```python
top_5 = top_corr_features.head(5).index.tolist()
sns.pairplot(df[top_5 + ['critical_temp']], diag_kind='kde')
```

```python
# checking if any feature has zero variance here
zero_var_cols = df.loc[:, df.std() == 0].columns
print("Columns with zero variance:", zero_var_cols.tolist())
```

```python
skewed_feats = df.drop(columns=['critical_temp']).skew().sort_values(ascending=False)
skewed_feats.head(10)
```
